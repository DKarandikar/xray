```python
import requests
import urllib.parse

# base_url='http://localhost:8118/api/apps?%s'
base_url='http://sociamnat.cs.ox.ac.uk:8080/api/apps?%s'

headers={ 'Accept': 'application/json' }
limit=10000
params = {"isFull": True, "limit":100, "appId":"com.miniclip.eightballpool"}

ss = urllib.parse.urlencode(params)
url = base_url % ss
url
```

```python
r = requests.get(url, headers=headers)
r
```
