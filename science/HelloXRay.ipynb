```python
import requests
import urllib.parse

# Make sure you're connected to the Oxford VPN before you try this
base_url='http://sociamnat.cs.ox.ac.uk:8080/api/apps?%s'

headers={ 'Accept': 'application/json' }
limit=10000

# these are the delicious parameters- first we'll start with 
# querying for ALL details of eightballpool
params = {"isFull": True, "limit":100, "appId":"com.miniclip.eightballpool"}
url = base_url % urllib.parse.urlencode(params)
url
```

```python
# that URL looks delicious! let's run it
r = requests.get(url, headers=headers)
r.json()
```
